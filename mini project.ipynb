```python
##Create a file

def create_file(filename):
    try:
        with open(filename,'w') as f:
            print(f"File '{filename}' created sucessfully.")
    except FileExistsError:
        print(f"File '{filename}' already exists.")
    except Exception as e:
        print(f"An error occured: {e}")

##Write a file

def write_to_file(filename, content):
    try:
        with open(filename, 'w') as f:
            f.write(content)
            print(f"Content written to '{filename}' sucessfully.")
    except FileNotFondError:
        print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"An error occured: {e}")
        
##Append a file
        
def append_to_file(filename, content):
    try:
        with open(filename, 'a') as f:
            f.write(content)
            print(f"Content appended to '{filename}' sucessfully.")
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"An error occured: {e}")

##Read a file

def read_to_file(filename):
    try:
        with open(filename, 'r') as f:
            content = f.read()
            print(f"Content of '{filename}' :\n{content}")
    except FileNoFoundError:
        print(f"File '{filename}' not found.")
    except Excepion as e:
        print(f"An error occured: {e}")

##Rename a file

import os
def rename_file(old_name,new_name):
    try:
        os.rename(old_name,new_name)
        print(f"File '{old_name}' renamed to '{new_name}' sucessfully.")
    except FileNotFoundError:
        print(f"File '{old_name}' not found.")
    except FileExistsError:
        print(f"File '{new_name}' already exists. Choose a different name.")
    except Exception as e:
        print(f"An error occured: {e}")

##Delete a file

import os
def delete_file(filename):
    try:
        os.remove(filename)
        print(f"File '{filename}' deleted successfully.")
    except FileNotFoundError:
        print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"An error occured: {e}")
        
##Creating menu

def file_manipulation_menu():
    while True:
        print("\n--- File Manipulation Menu ---")
        print("1. Create File")
        print("2. Write to File")
        print("3. Append to File")
        print("4. Read File")
        print("5. Rename File")
        print("6. Delete File")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            os.chdir("C:\\Users\\DELL\\Desktop\\New folder")
            f=open("my_new_file.txt","w")
            f.close()
            
        elif choice == '2':
            f=open("my_new_file.txt","w")
            f.write("File handling is the Project...")
            f.close()
           
        elif choice == '3':
            f=open("my_new_file.txt","a")
            f.write("\nIt is the first mini project...")
            f.close()
            
        elif choice == '4':
            f=open("my_new_file.txt","r")             
            content =f.read()
            print(content)
            f.close()
           
        elif choice == '5':
            import os
            os.rename("my_new_file.txt","my_document.txt")
           
        elif choice == '6':
            os.remove("my_document.txt")
            
        elif choice == '7':
            print("Existing file manipulation program.")
            break
        else:
            print("Invalid choice.Please enter a number between 1 and 7.")

#run the menu
file_manipulation_menu()
        
```
